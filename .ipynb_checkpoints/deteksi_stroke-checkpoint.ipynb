```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/healthcare-dataset-stroke-data.csv')
```

```python
df[['gender', 'id']].groupby(['gender']).count()
```

```python
# Hapus data gender Other
df = df[df['gender']!='Other']
df.head(5)
```

```python
# df.isnull().sum()
# 
```

```python
# hitung rata-rata bmi pasien stroke untuk digantikan ke data bmi pasien stroke yg kosong
rata_rata_bmi_stroke = df[df['stroke']==1]['bmi'].mean()
rata_rata_bmi_stroke
```

```python
# hitung rata-rata bmi pasien nonstroke untuk digantikan ke data bmi pasien stroke yg kosong
rata_rata_bmi_nonstroke =  df[df['stroke']==0]['bmi'].mean()
rata_rata_bmi_nonstroke
```

```python
df.loc[df['stroke']==1, 'bmi'] = df.loc[df['stroke']==1, 'bmi'].fillna(rata_rata_bmi_stroke)
df.loc[df['stroke']==0, 'bmi'] = df.loc[df['stroke']==0, 'bmi'].fillna(rata_rata_bmi_nonstroke)
```

```python
df[['stroke','id']].groupby(['stroke']).count().reset_index()
```

```python
all_df = df.copy()

stroke = df[df['stroke']==1]
non_stroke = df[df['stroke']==0]
```

```python
def plot_pie(column, title="All"):
    fig,axs = plt.subplots(1,1)
    data = all_df[column].value_counts()
    plt.pie(data,autopct='%1.2f%%',labels=data.index)
    plt.title(title)
    plt.show()
    
def plot_hist(column, title="all"):
    plt.hist(all_df[column],density=True)
    plt.title(title)
    plt.show()

def plot_bar(column, sort=False, title="all"):
    if sort:
        data_all = all_df[column].value_counts().sort_index()
    else:
        data_all = all_df[column].value_counts()
    plt.bar(data_all.index,data_all)
    plt.title(title)
    plt.show()
    
def plot_bar_compare(column, sort=False):
    if sort:
        data_churned = stroke[column].value_counts().sort_index()
        data_nonchurned = non_stroke[column].value_counts().sort_index()
    else:
        data_churned = stroke[column].value_counts()
        data_nonchurned = non_stroke[column].value_counts()
    
    fig,axs = plt.subplots(2,1)
    plt.subplots_adjust(left=0, bottom=0, right=1, top=2, wspace=0, hspace=0.2)
    axs[0].bar(data_nonchurned.index,data_nonchurned)
    axs[0].title.set_text('No Stroke')
    axs[1].bar(data_churned.index,data_churned)
    axs[1].title.set_text('Has Stroke')
    plt.show()

def plot_hist_compare(column, bins=5):
    plt.hist([non_stroke[column], stroke[column]] , color=['c','r'])
    plt.legend(('No Stroke', 'Has Stroke'))
    plt.show()
    
def plot_pie_compare(column):
    data_churned = stroke[column].value_counts()
    data_nonchurned = non_stroke[column].value_counts()
    
    fig,axs = plt.subplots(2,1)
    plt.subplots_adjust(left=0, bottom=0, right=1, top=2, wspace=0, hspace=0.2)
    axs[0].pie(data_nonchurned,autopct='%1.2f%%',labels=data_nonchurned.index)
    axs[0].title.set_text('No Stroke')
    axs[1].pie(data_churned,autopct='%1.2f%%',labels=data_churned.index)
    axs[1].title.set_text('Has Stroke')
    plt.show()

def plot_boxplot(column, title=""):
    sns.boxplot(x="stroke", y=column, palette=["c", "r"],
            hue="stroke",  data=all_df).set_title(title, fontsize=15)

def check_median(column):
    data_churned = stroke[column].describe()
    data_nonchurned = non_stroke[column].describe()
    print('No Stroke: {}'.format(data_nonchurned['50%']))
    print('Has Stroke: {}'.format(data_churned['50%']))

def check_most(column):
    data_churned = stroke[column].value_counts()
    data_nonchurned = non_stroke[column].value_counts()
    print('No Stroke: {}'.format(data_nonchurned.index[0]))
    print('Has Stroke: {}'.format(data_churned.index[0]))
```

```python
# Data Preprocessing
X = df.copy()
y = X['stroke']

X = X.drop(['stroke', 'gender', 'work_type','Residence_type', 'ever_married', 'smoking_status','id'], axis=1)
```

```python
# Split data ke train dan test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=1234)

X_smote, y_smote = sm.fit_resample(X_train, y_train)

print(f'''Shape of X before SMOTE: {X.shape}
Shape of X after SMOTE: {X_smote.shape}''')

print('\nBalance of positive and negative classes (%):')
y_smote.value_counts(normalize=True) * 100
```

```python
# # Scalling
# from sklearn.preprocessing import StandardScaler

# scale = StandardScaler()

# X_train = scale.fit_transform(X_train)
# X_test = scale.transform(X_test)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_smote = sc.fit_transform(X_smote)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf

# Membuat model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_dim=X_train.shape[1], activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Melatih model dengan data training
model.fit(X_smote, y_smote, epochs=50, batch_size=64)
```

```python
# Evaluasi model pada data test
accuracy = model.evaluate(X_test, y_test)[1]
print(f'Accuracy on test data: {accuracy}')
```
